```python
#importing required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from apyori import apriori
```

```python
dataFrame=pd.read_csv('Market_Basket_Optimisation.csv')           #using pandas to read the data
dataFrame
```

```python
dataFrame.shape
```

```python
dataFrame.info()
```

```python
#getting list of transactions from dataset
transactions=[]
for i in range(0,7500):
    transactions.append([str(dataFrame.values[i,j]) for j in range(0,20)]) #adding / appending elements to the end of list
    dataFrame.head
```

```python
dataFrame.head()
```

```python
#training apriori algorithm on the dataset
rule_list=apriori(transactions, min_support=0.004, min_confidence=0.3, min_lift=3, min_length=2)
```

```python
#visualising the list of rules 
results=list(rule_list)
for i in results:
    print('/n')
    print(i)
    print('******************************')
```
